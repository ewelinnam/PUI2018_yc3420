### use groupby to count the building per LandUse code

```python
bkgrouped = bk.groupby('LandUse').count()
bkgrouped = bkgrouped.reset_index()
bkgrouped
```

```python
mngrouped = mn.groupby('LandUse').count()
mngrouped = mngrouped.reset_index()
mngrouped
```

### convert the landuse code to integer

```python
bkgrouped['LandUse'] = bkgrouped['LandUse'].astype(int)
bkgrouped
```

```python
mngrouped['LandUse'] = mngrouped['LandUse'].astype(int)
mngrouped
```

### merge and rename the dataframe

```python
res = pd.merge(bkgrouped, mngrouped, on = 'LandUse')
res
```

```python
res = res.rename(columns = {'ZipCode_x' : 'bk', 'ZipCode_y': 'mn'})
res
```

### find online the LandUse data dictionary and replace code w the verbose description of it; set up land use values

```python
code_name = ['One &Two Family Buildings','Multi-Family Walk-Up Buildings',\
            'Multi-Family Elevator Buildings','Mixed Residential & Commercial Buildings',\
            'Commercial & Office Buildings','Industrial & Manufacturing','Transportation & Utility',\
            'Public Facilities & Institutions','Open Space & Outdoor Recreation','Parking Facilities',\
            'Vacant Land']
#for i in res['LandUse']:
#    print (i, code_name[i-1])
```

```python
for i in range(len(res)):
    res['LandUse'] = res['LandUse'].replace(i,code_name[i])
    
res
```

### create a feature (a column, a series) with the ratio of number of buildings in BK vs MN and find out in which LandUse class Manhattan has more buildings than BK

```python
res['luratio'] = res['mn'] / res['bk']
res
```

```python
res['Manhattan wins'] = ''
for i in range(len(res)):
    if res['luratio'][i] > 1:
        res['Manhattan wins'][i] = 'True'
    else:
        res['Manhattan wins'][i] = 'False'
```

```python
res
```

 In the analysis above, we could see that only in multi-family elecator buildings category, the number of Manhattan is larger than Brooklyn. However, this conclusion is not aligned with common sense. From my prospective, this error was introduced because we ignored the difference of the building size in bk/mn (which is, a small house and a residential building which could accommodate 3000 people) and only count on the absolute number of the building itself. Another reason could be, each Landuse region in Manhattan is much smaller than that in brooklyn. Thus, the absolute number of brooklyn buildings is larger than that of Manhattan.
